```python
"""
docstring:

A) “The heart break” is a Text Adventure game which simulates real life heart break situation of a human being and a series of un/fortunate decisions that led him to his final destiny. The game consist of a 3 stage game, where you are ask to different questions and depending your answer the different paths that the game will take you too.
     o Stage_0= Intro
     o Stage_1 = A bench in the park
     o Stage_2 = Party with Roxanne
     o Stage_3 = Outside your house

B) Known Bugs and/or Errors:

     o Spelling and syntax errors.
"""

#!# PLEASE RUN THIS FIRST #!#


#We run this to create an easy way to add colors to our game
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    
import time
import random
import pandas as pd

```

```python

#Game intro - Press Start
def press_start():
    
    #Color blue - Press start message and input
    print(f"{bcolors.OKBLUE} Please type start \n")
    start = input(prompt = ">>>  ")
    start = start.lower()
    
    # Conditional input - Press Start
    
    if "start" in start:
        
        # if start ---> Start game!
        stage_0()

    elif "no" in start:
        #if no ---> take me out
        fail_1() 
    
    else:
        #incorrect entry
        print(f"{bcolors.WARNING}Invalid entry. Please try typing >start< again.")
        press_start()
        
def fail_1():
    #print statment whe "no" at press start
    print(f"""{bcolors.HEADER} Run the code again if you want to play!!""")
    exit()
    

def stage_0():
    
    #Visual Effect for the into
    telon = 2

    while telon > 0:
        print(".")
        time.sleep(0.4)
        print(".")
        time.sleep( 0.4 )
        telon -=1
    user_name()

#Fail 2  - Wrong anwswer at the park 1
def fail_2(user):
    print(f"""
    _______________________________________________________________________
    
    {bcolors.WARNING}¡Oh no! You made the wrong choice.
    
    {bcolors.HEADER} The old man is pointing you with a gun !!
    You are completely stunned and you don't react. The old man gets nervous, so he shoots you, takes your money 
    and run!! You are lying in the floor, bleeding out and still you are thinking of you Ex.

    {bcolors.BOLD}{bcolors.OKCYAN}G A M E    O V E R   !!  
    _______________________________________________________________________ 
     
    """)
    play_again_1(user)   

#Fail 3  - Wrong anwswer at the park 2 
def fail_3(user):
    #Wrong choice! 
    print(f"""
    _______________________________________________________________________
    
    {bcolors.WARNING} :S You made the wrong choice!
    
    {bcolors.HEADER} The old man goes away. You stay in the bench,
    and after a long inner reflexion, you decide to go back home with your heartache,
    eat ice cream and watch "Love Actually" while crying in bed. 
     
    {bcolors.BOLD}{bcolors.OKCYAN}G A M E    O V E R   !!  
    _______________________________________________________________________ 
     
    """)
    
    play_again_1(user)

def first_interaction(user): #First interaction
        t=0.8
        print(f"Old man:{bcolors.HEADER} May I sit with you? (yes/no)")
        
        answer1 = input(prompt="")
        answer1 = answer1.casefold()
        
        if "yes" in answer1:
            #Old man sits next to you
            time.sleep(t)
            stage_1(user)

        elif "no" in answer1:
            #Old man is offended
            print(f"{bcolors.OKBLUE}Old man:{bcolors.HEADER} You don't have to be rude... I will be quick then. Your money or your life!!!")
            time.sleep(t)
            fail_2(user)

        else:
            #You dont respond, old man ask for money
            print(f"{bcolors.OKBLUE}Old man:{bcolors.HEADER}Sorry, I don't mean to bother....")
            time.sleep(t)
            first_interaction(user)
    
    
def user_name():
 #User input - This way you'll have a name in the game   
    user = input(prompt=f"{bcolors.ENDC} Could you please write your name? ")
    user = user.capitalize()
    
    #Confirm user name
    print(f"You typed {user}, is that correct? (yes / no)")
    
    userconf = input(prompt=">>>")
    userconf = userconf.casefold()
    
    #conditional user confirmation
    if "yes" in userconf:
        print(f"Thanks {user}!")
        
        #Visual Effect
        telon = 3
        while telon > 0:
            print(".")
            time.sleep(0.3)
            print(".")
            time.sleep( 0.3 )
            telon -=1
          
        #Game Intro
        t=0.8
        time.sleep(t)
        print(".... Long.")
        time.sleep(t)
        print("......... Long.")
        time.sleep(t)
        print ("............ Long time ago.") 
        time.sleep(t)
        print("Well, maybe around a year... ") #joke
        time.sleep(t+0.3)
        print("Nevermind, it was a cold november afternoon when I received that hurtful message.")
        time.sleep(t)
        print(f"The message that starts with '{user} we need to talk'.")
        time.sleep(t)
        print("There I was, sitting in that bench in the middle of the park.")
        time.sleep(t)
        print("Around me, kids laughing, family reunions, lovely dogs chasing their tail.")
        time.sleep(t)
        print("Everyone so happy, but me.")
        time.sleep(t) 
        print("While seconds turned it minutes, and minutes turn into hours.") 
        time.sleep(t) 
        print("I could not think on anything else but the terrible feeling of a broken heart")
        time.sleep(t)
        print("Suddenly, someone approaches me.")
        time.sleep(t)
        
        #Visual Effect
        telon = 3
        while telon > 0:
            print(".")
            time.sleep(0.3)
            print(".")
            time.sleep( 0.3 )
            telon -=1
            
        first_interaction(user)
        
    #User Name Confirmation Condicional    
    elif "no" in userconf:
        #If no ---> retype it
        print(f"{bcolors.WARNING}No problem, let's try again.\n")
        user_name()
        
    else:
        #else ---> Invalid entry
        print(f"{bcolors.WARNING}Invalid entry. Please try typing again.\n")
        user_name()

    


#Play again option
def play_again_1(user):
    play = input(" Want to play again? (yes/no) ")
    play = play.casefold()
    
    if "yes"in play:
        print("Great! Lets go back to the beginnig.")
        input("Press Enter when ready")
        first_interaction(user)
    
    elif "no" in play:
        fail_1() 
    
    else:
        print(f"{bcolors.WARNING}Invalid entry. Please try again.")
        play_again_1(user)
        



```
